# Matplotlib
#### This is for *SIMPLE* matplotlib plotting only. For beauty and ease of use, use other tools. 

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
```

#### Plot Format: 

Basic Steps: 

1. Plot (x, y) and specify type of charts: 
    
        Line Chart: plt.plot(x, y)
        Scatter plot: plt.plot(x, y, 'o')
    
2. To designate x and y axis: 
        plt.axis([x_lb, x_ub, y_lb, y_ub])

3. To show plot: 
        plt.show()

#### Formatting - Line Chart: 
1. Line width Control: 
        plt.plot(x, y, linewidth = 2.0)
2. Line style: linestyle = '-'
        '-';      '--';      '-.' ;      ':' ;
        
#### Formatting - Scatter Plots: 
1. Color: 
         red: r; 
         green: g; 
         blue: b; 
2. Dot type: 
         dash: -; 
         dash and dot: -.; 
         very small dot: :; 
         bit dot: o; 
         square: s; 
         triangle: ^; 
         Others: '|', '+'; 
3. Marker size:     markersize = 2.0

### Plot multiple charts


```python
x1 = np.arange(0, 40, 2)
y1 = np.arange(10, 410, 20)
x2 = np.arange(1, 41, 2)
y2 = np.arange(20, 420, 20)


np.random.seed(999)
x3 = 100 + 20 * np.random.randn(10000)

x4 = np.arange(1, 60, 3) #20 steps
y4 = 40 + x4 *  np.random.randn(20)
```

```python
# Step 1. Set up layout.   In this example, I want to set up 4 plots. 2 x 2.
# You can use ax1 = plt.subplot() or omit ax1 =  and use plt.subplot directly. But I prefer use the ax1 method. 

# Note: from this line down, the figure size is set. 
from matplotlib.pyplot import figure #set figure size / chart size
figure(figsize=(16, 8), dpi=80, facecolor='grey', edgecolor='k') 
plt.suptitle('This is the overall title', fontsize=20)

# ax1: an easy example. 
ax1 = plt.subplot(221) # row 1 col 2, Index 1
ax1 = plt.plot(x1, y1, 'g', linestyle = '-.', linewidth = 3.0)

# ax2: One level up. Line and scatter plot 
ax2 = plt.subplot(222) # row 1 col 2, Index 2
# dual plots below. Line and line. 
ax2 = plt.plot(x2, y2, '+', color = 'red',  markersize = 5.0) # Order matters! Makrer style first then color. 
ax2 = plt.plot(x2-10, y2+10, color = 'g', linestyle = '-', linewidth = 3.0) 
# Step 2: Add formatting items,i.e. title, text, grid, label name, axis range, etc. 
# You can omit all the 'ax2=' below but I want to keep it just to get the display clear. 
ax2 = plt.title("2nd Plot's Title")
ax2 = plt.text(15, 150, 'Say something')
ax2 = plt.grid(True)
ax2 = plt.xlabel('x axis name', fontsize =15)
ax2 = plt.ylabel('y axis name')
ax2 = plt.axis([-20, 40, -10, 440])

# plt.delaxes(ax2) #delete ax2 from the figure
# plt.subplot(ax2) #add ax2 to the figure again

ax3 = plt.subplot(223) # add subplot
n, bins, patches = plt.hist(x3, 50, normed = 1, facecolor = 'magenta', alpha = 0.75)
ax3 = plt.axis ([40, 160, 0, 0.03])

ax4 = plt.subplot(224)
ax4 = plt.plot(x4, y4, '^', color = 'red',  markersize = 4.5 )
ax4 = plt.bar(x4, y4, width=0.8, bottom=None, align='center', color = 'black')
ax4 = plt.ylim(0)

# step 3: show
plt.show()
```

```python
fig = plt.figure(figsize = (10, 10), dpi=80, facecolor='grey', edgecolor='k')
fig.suptitle('Ugly - But best way to demo codes', size = 20.0)

ax1 = fig.add_subplot(111)
ax1.plot(x1, y1, color = 'red', label = 'legend 1')
ax1.set_ylim(-30, 400)
ax1.set_xlabel('shared x label', size = 14.0, color = 'purple')
ax1.set_ylabel('y1', size = 14.0)
ax1.grid(True)
ax1.tick_params(direction='out', length=60, width=10, colors='yellow',
                grid_color='brown', grid_alpha=0.2, size = 50, rotation=45, labelsize = 5)
ax1.xaxis.set_tick_params(labelsize=30)

ax2 = ax1.twinx()
ax2.bar(x4, y4, color = 'green' , label = 'legend 1')
ax2.set_ylim(-30, 400)
ax2.set_ylabel('y4', size = 14.0)

ax3 = ax1.twinx()
ax3.plot(x4, y4, 'o', color = 'magenta')
ax3.set_ylim(-30, 400)

plt.xticks(np.arange(0, max(x4)+5, 4))

h1, l1 = ax1.get_legend_handles_labels()
print(h1)
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(handles=h1+h2, labels=l1+l2, loc='upper right', frameon = False)
```

### Plot Colors

Use imshow() 
See below method from 
https://matplotlib.org/gallery/axes_grid1/simple_colorbar.html#sphx-glr-gallery-axes-grid1-simple-colorbar-py

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

ax1 = plt.subplot(121)
im1= ax1.imshow(np.arange(100).reshape((10, 10)))

# create an axes on the right side of ax. The width of cax will be 5%of ax and the padding between cax and ax will be fixed at 0.05 inch.
divider = make_axes_locatable(ax1)
cax = divider.append_axes("right", size="5%", pad=0.05)
ax1 = plt.colorbar(im1, cax=cax)

ax2 = plt.subplot(122)
im2 = ax2.imshow(np.arange(100).reshape((10, 10)))
ax2 = plt.colorbar(im2)
```
