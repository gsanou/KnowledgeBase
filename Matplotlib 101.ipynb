```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.pyplot import figure
```

There are two ways of plotting.  
    1. plt.plot() method. 
    2. Class method (object-oriented). 


```python
# Using pyplot.plt method to plot. Quick way to plot. 

plt.subplot(211)
plt.plot([x for x in range(0, 25, 1)], [y**2 for y in range(0, 25, 1)])
plt.plot([x for x in range(0, 25, 1)], [y for y in range(0, 25, 1)])
plt.title('square numbers')

plt.subplot(212)
plt.plot([x for x in range(0, 25, 1)], [y**3 for y in range(0, 25, 1)])
plt.title('cubed numbers')

plt.show()
```

```python
# Using class method to plot. Allows more controls

fig, ax = plt.subplots(2) #two subplots # fig is a class. ax is a numpy array. 
ax[0].plot([x for x in range(0, 25, 1)], [y**2 for y in range(0, 25, 1)])
ax[0].plot([x for x in range(0, 25, 1)], [y for y in range(0, 25, 1)])
ax[1].plot([x for x in range(0, 25, 1)], [y**3 for y in range(0, 25, 1)])
```

#### Plot Format: 

Basic Steps: 

1. Plot (x, y) and specify type of charts: 
    
        Line Chart: plt.plot(x, y)
        Scatter plot: plt.plot(x, y, 'o')
    
2. To designate x and y axis: 
        plt.axis([x_lb, x_ub, y_lb, y_ub])

3. To show plot: 
        plt.show()

#### Formatting - Line Chart: 
1. Line width Control: 
        plt.plot(x, y, linewidth = 2.0)
2. Line style: linestyle = '-'
        '-';      '--';      '-.' ;      ':' ;
        
#### Formatting - Scatter Plots: 
1. Color: 
         red: r; 
         green: g; 
         blue: b; 
2. Dot type: 
         dash: -; 
         dash and dot: -.; 
         very small dot: :; 
         bit dot: o; 
         square: s; 
         triangle: ^; 
         Others: '|', '+'; 
3. Marker size:     markersize = 2.0

```python
# Step 1. Set up layout.   In this example, I want to set up 4 plots. 2 x 2.
# You can use ax1 = plt.subplot() or omit ax1 =  and use plt.subplot directly. But I prefer use the ax1 method. 

# Note: from this line down, the figure size is set. 
fig = plt.figure(figsize=(16, 8), dpi=80, facecolor='grey', edgecolor='k') 
plt.suptitle('This is the overall title. The one', fontsize=20)

# ax1: an easy example. 
ax1 = fig.add_subplot(221) # 2x2, index 1
ax1 = plt.plot(x1, y1, 'g', linestyle = '-.', linewidth = 3.0)

# ax2: A bit complicated. Line and scatter plot 
ax2 = fig.add_subplot(222) # row 1 col 2, Index 2
# dual plots below. Line and line. 
ax2 = plt.plot(x2, y2, '+', color = 'red',  markersize = 5.0) # Order matters! Makrer style first then color. 
ax2 = plt.plot(x2-10, y2+10, color = 'g', linestyle = '-', linewidth = 3.0) 
# Step 2: Add formatting items,i.e. title, text, grid, label name, axis range, etc. 
# You can omit all the 'ax2=' below but I want to keep it just to get the display clear. 
ax2 = plt.title("2nd Plot's Title")
ax2 = plt.text(15, 150, 'Say something')
ax2 = plt.grid(True)
ax2 = plt.xlabel('x axis name', fontsize =15)
ax2 = plt.ylabel('y axis name')
ax2 = plt.axis([-20, 40, -10, 440])
# ax2 = plt.legend(bbox_to_anchor=(1.1, 1.05))
ax2 = plt.legend('legend name', loc = 'lower left')
# plt.delaxes(ax2) #delete ax2 from the figure
# plt.subplot(ax2) #add ax2 to the figure again

ax3 = fig.add_subplot(223) # 2x2, index 3
n, bins, patches = plt.hist(x3, 50, normed = 1, facecolor = 'magenta', alpha = 0.75)
ax3 = plt.axis ([40, 160, 0, 0.03])

ax4 = fig.add_subplot(224)
ax4 = plt.plot(x4, y4, '^', color = 'red',  markersize = 4.5 )
ax4 = plt.bar(x4, y4, width=0.8, bottom=None, align='center', color = 'black')
ax4 = plt.ylim(0)

# step 3: show
plt.show()

# step 4: Save figure to local drive
fig.savefig('myfig.png') 
# other common formats available: jpeg, jpg, pdf, tif, tiff, rgba
fig.canvas.get_supported_filetypes()
```

### Plot Colors / Images

Use imshow() 
See below method from 
https://matplotlib.org/gallery/axes_grid1/simple_colorbar.html#sphx-glr-gallery-axes-grid1-simple-colorbar-py

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

ax1 = plt.subplot(121)
im1= ax1.imshow(np.arange(100).reshape((10, 10)))

# create an axes on the right side of ax. The width of cax will be 5%of ax and the padding between cax and ax will be fixed at 0.05 inch.
divider = make_axes_locatable(ax1)
cax = divider.append_axes("right", size="5%", pad=0.05)
ax1 = plt.colorbar(im1, cax=cax)

ax2 = plt.subplot(122)
im2 = ax2.imshow(np.arange(100).reshape((10, 10)))
ax2 = plt.colorbar(im2)
```

```python
fig = plt.figure(figsize = (18,12))
plt.suptitle('overall title', fontsize = 15)
ax1 = fig.add_subplot(221)
df[['x1', 'y1']].plot.line(x = 'x1', y = 'y1',  marker = 'o', linestyle = '-.', color = 'red', ax = ax1)
df[['x4', 'y4']].plot.bar(x = 'x4',y = 'y4',   color = 'blue', ax = ax1 )

ax2 = fig.add_subplot(222)
df[['x5', 'y5']].plot.barh(x = 'x5', y = 'y5', ax = ax2)

ax3 = fig.add_subplot(223)
# df[[ 'y4']].plot.box( y = 'y4', ax = ax3)
df.boxplot(column = ['y1',  'y2'], ax = ax3, grid = False)

ax4 = fig.add_subplot(224)
df[['x2', 'y4']].plot.scatter(x = 'x2', y = 'y4', marker = 's', color = 'magenta', ax = ax4)
```
